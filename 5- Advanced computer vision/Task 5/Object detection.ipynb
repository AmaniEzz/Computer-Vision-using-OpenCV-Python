## Detect an object with OpenCV-Python


We will do object detection in this article using something known as haar cascades. **Haar Cascade** is a machine learning-based approach where a lot of positive and negative images are used to train the classifier.

```python
import cv2 
from matplotlib import pyplot as plt 
  

# Opening image 
img = cv2.imread("recognize_stop_sign/image.jpg") 
  
# OpenCV opens images as BRG but we want it as RGB and as grayscale  
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
  
# Creates the environment of the picture and shows it 
plt.imshow(img_rgb) 
plt.show() 
```

### We will use the detectMultiScale() function of OpenCV to recognize big signs as well as small ones:

```python
# Use minSize to avoid extra-small dots that would look like STOP signs.

stop_data = cv2.CascadeClassifier('recognize_stop_sign/stop_data.xml') 
  
signs = stop_data.detectMultiScale(img_gray,  minSize =(20, 20)) 

# Check if any signs exits
amount_found = len(signs)

# Don't do anything if there's no sign 
if amount_found != 0: 
    
    # There may be more than one sign in the image 
    for (x, y, width, height) in signs: 
          
        # We draw a green rectangle around every recognized sign 
        cv2.rectangle(img_rgb, (x, y),  (x + height, y + width),  (0, 255, 0), 5) 
          

plt.imshow(img_rgb) 
plt.show() 
```
